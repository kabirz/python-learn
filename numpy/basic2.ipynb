```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([2, 3])
c = np.array([3, 4, 5, 6])
ax, bx, cx = np.ix_(a, b, c)
print('#'*40)
print(ax, ax.shape)
print('#'*40)
print(bx, bx.shape)
print('#'*40)
print(cx, cx.shape)
print('#'*40)

res = ax + bx * cx
print(res, res.shape)
print('#'*40)
print(res[3, 1, 2], a[3] + b[1] * c[2])
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 5, 21)
y = np.linspace(0, 10, 21)
plt.plot(x, y, 'purple') # line
plt.plot(x, y, 'o'); # dots
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis');
```
