```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

def show_image(img):
	plt.axis('off')
	plt.gca().xaxis.set_major_locator(plt.NullLocator())
	plt.gca().yaxis.set_major_locator(plt.NullLocator())
	plt.subplots_adjust(top=1, bottom=0, left=0, right=1, hspace=0, wspace=0)
	plt.margins(0, 0)
	plt.imshow(img)
	plt.show()
```

```python
a = np.fromfile('img.grbg', dtype=np.uint16).reshape(240, 320)
a = a/4
a = a.astype(np.uint8)
J = cv.cvtColor(a, cv.COLOR_BAYER_GR2RGB)

show_image(J)
```
